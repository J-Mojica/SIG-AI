# Zero One Classifier

## Imports

```python
import MNIST
import perceptron
import numpy as np
import matplotlib.pyplot as plt
```

## Load the data

```python
(x_train, y_train), (x_test, y_test) = MNIST.load_zeros_and_ones()
```

### Visualizing an example

```python
def plot_image(x):
    '''Plot an image'''
    plt.imshow(MNIST.vector_to_matrix(x), cmap='gray')
    plt.show()
```

```python
plot_image(x_train[0])
```

### Normalize data to 0-1 Range

This shouldn't change the way we classify the data

```python
x_train, x_test = x_train/255.0, x_test/255.0
```

## Train the model

Using the perceptron algorithm we coded

```python
theta, theta_0 = perceptron.perceptron(x_train, y_train, 1)
```

We make a function to which we can pass a new image, and get a prediction

```python
model_predict = lambda x: np.sign(theta.T @ x + theta_0)
```

```python
plot_image(x_test[500])
```

```python
model_predict(x_test[400]) # Let's go!
```

```python
for x in x_test:
    if(model_predict(x) != 1):
        print("We did something different than one")

print(max(theta))
```

# Evaluating our model

```python
accuracy = 0.0
print(x_test.shape)
print(len(y_test))
for i in range(len(x_test)):
    prediction = model_predict(x_test[i])
    if(prediction == y_test[i]):
        accuracy += 1
accuracy /= len(x_test)
print(accuracy)
```
